```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./train/train.csv')
```

```python
data.head()
```

```python
print(len(data))
```

```python
sns.set({'ytick.major.width': 0.8, 'ytick.major.size': 8.0,})
```

```python
fig = plt.figure(figsize=(30,40))
sns.set(font_scale=5)
sns.countplot(y='label', data=data)
```

```python
fig, ax = plt.subplots(2,1,figsize=(30,80))
sns.set(font_scale=5)
sns.countplot(y='label', data=data[:3247], ax = ax[0])
sns.countplot(y='label', data=data[3247:3247*2], ax = ax[1])
```

```python
plt.xticks(rotation = 90 )
sns.countplot(x='label', data=data)
```
